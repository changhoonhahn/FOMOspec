```python
import numpy as np
# --- fomospec ---
from FOMOspec import util as UT
from FOMOspec import fitters as Fitters
```

```python
import matplotlib as mpl 
import matplotlib.pyplot as pl 
mpl.rcParams['text.usetex'] = True
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['axes.linewidth'] = 1.5
mpl.rcParams['axes.xmargin'] = 1
mpl.rcParams['xtick.labelsize'] = 'x-large'
mpl.rcParams['xtick.major.size'] = 5
mpl.rcParams['xtick.major.width'] = 1.5
mpl.rcParams['ytick.labelsize'] = 'x-large'
mpl.rcParams['ytick.major.size'] = 5
mpl.rcParams['ytick.major.width'] = 1.5
mpl.rcParams['legend.frameon'] = False
%matplotlib inline
```

```python
prosp = Fitters.Prospector()
```

```python
lam = np.linspace(3000, 1e4, 1e4)
tt = {'logzsol': -0.5, 
      'dust2': 0.3, 
      'tau': 3., 
      'tage': 12., 
      'mass': 1e9}
flux, _, _ = prosp.model(lam, 0.1, tt)
```

```python
fig = plt.figure(figsize=(15,5))
sub = fig.add_subplot(111)
tt = {'logzsol': -0.5, 
      'dust2': 0.3, 
      'tau': 3., 
      'tage': 12., 
      'mass': 1e9}
flux, _, _ = prosp.model(lam, 0.1, tt)
sub.plot(lam, flux)
tt = {'logzsol': -0.5, 
      'dust2': 0.3, 
      'tau': 3., 
      'tage': 12., 
      'mass': 5e9}
flux, _, _ = prosp.model(lam, 0.1, tt)
sub.plot(lam, flux, ls=':')
sub.set_xlim([3e3, 1e4])
sub.set_ylim([0., 10e-8])
```

```python
fig = plt.figure(figsize=(15,5))
sub = fig.add_subplot(111)
for logzsol in np.linspace(-1., 0., 5):
    tt = {'logzsol': logzsol, 'dust2': 0.3, 'tau': 3., 'tage': 12., 'mass': 1e10}
    flux, _, _ = prosp.model(lam, 0.1, tt)
    sub.plot(lam, flux, label=r"$\log (Z/Z_\odot)="+str(logzsol)+"$")
sub.legend(loc='upper left', fontsize=20)
sub.set_xlim([3e3, 1e4])
sub.set_ylim([0., 10e-8])
```

```python
fig = plt.figure(figsize=(15,5))
sub = fig.add_subplot(111)
for d2 in [0.1, 0.3, 0.5, 0.7]:
    tt = {'logzsol': -0.5, 'dust2': d2, 'tau': 3., 'tage': 12., 'mass': 1e10}
    flux, _, _ = prosp.model(lam, 0.1, tt)
    sub.plot(lam, flux, label='dust2='+str(d2))
sub.legend(loc='upper left', fontsize=20)
sub.set_xlim([3e3, 1e4])
sub.set_ylim([0., 10e-8])
sub.set_title('optical depth at 5500AA', fontsize=20)
```

```python
fig = plt.figure(figsize=(15,5))
sub = fig.add_subplot(111)
for tau in [0.1, 1., 3., 10.]:
    tt = {'logzsol': -0.5, 'dust2': 0.3, 'tau': tau, 'tage': 12., 'mass': 1e10}
    flux, _, _ = prosp.model(lam, 0.1, tt)
    sub.plot(lam, flux, label=r'$\tau$='+str(tau))
sub.legend(loc='upper left', fontsize=20)
sub.set_xlim([3e3, 1e4])
sub.set_ylim([0., 10e-8])
```
