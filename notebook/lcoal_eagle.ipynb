```python
fig = plt.figure(figsize=(10,5*3))
np.random.seed(1)
for ii, i in enumerate(np.random.choice(np.arange(spec.shape[0]), 3)):
    print(i)
    sub = fig.add_subplot(3,1,ii+1)
    sub.plot(lam, spec[i,:])
    sub.set_xlabel('Wavelength [$\lambda$]', fontsize=25)
    sub.set_xlim([3000., 10000.])
```
